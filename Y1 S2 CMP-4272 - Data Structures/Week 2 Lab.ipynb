3. Previously you have already written python functions for union and intersection of two sets.
Write the two functions to compute the union and intersection of sets making use of the following:
```
•	set() creates an empty set,
•	y in my_set checks if the set contains y,
•	len(my_set) gets the size of the set,
•	my_set.add(y) inserts y into the set if not already present,
•	my_set.remove(y) removes y from the set if present,
•	my_set.clear() clears the set,
•	for k in my_set: iterates over the set.
```







````python
# 3. Your code goes in here ( for union and intersection functions)
def sampUnion(set1, set2):
  result = set()
  for x in set1:
    result.add(x)
  for y in set2:
    result.add(y)
  print(result)

def sampIntersection(set1, set2):
  result = set()
  for x in set1:
    if x in set2:
      result.add(x)
  print(result)

z = {"one", "seven", "two"}
x = {"four", "one", "nine"}

sampUnion(z, x)
sampIntersection(z, x)
````

5. Making use of both Stack and Queue (updated), you are required to write a python function that checks if a given number is pallindrome or not.

Example of Pallindrome numbers: 1221, 550055, 1111, 41814 , etc.
[ Pallindrome - reads the same backwards as forwards]


*   
[Hint : Push each digit on the stack. Similarly Enqueue each digit in the Queue.
Then compare values returned by push and pop operations]

````python
# 5. Check for a pallindrome (using a combiation of stack and queue)

def palindromeCheck(data):
  sample = []
  reversed = ''


  for elem in str(data):
    enqueue(sample, elem)

  while checkIfEmpty(sample) != "Empty":
    reversed = reversed + pop(sample)

  if str(data) == reversed:
    print(data, "is a plaindrome")
  else:
    print(data, "is not a a palindrome")

palindromeCheck(1221)
palindromeCheck(550055)
palindromeCheck(9172938123123)
````
